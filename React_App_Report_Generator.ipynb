```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import data
coefficients = pd.read_csv("C:\\Users\\Bajo\\OneDrive\\Desktop\\GMU GRA Project\\coefficients-REMOVE.csv")
```

```python
coefficients['strata'] = coefficients.apply(lambda row: f"{row['cgrp']}:{row['ctype']}:{row['code']}", axis=1)
```

```python
# Make it so that antidep only has the first letter capitalized

coefficients['antidep'] = coefficients['antidep'].str.capitalize()
```

```python
# Mapping of antidepressants to their NIH links

nih_links = {
    'Amitriptyline': 'https://www.ncbi.nlm.nih.gov/books/NBK537225/',
    'Bupropion': 'https://www.ncbi.nlm.nih.gov/books/NBK470212/',
    'Citalopram': 'https://www.ncbi.nlm.nih.gov/books/NBK482222/',
    'Desvenlafaxine': 'https://www.ncbi.nlm.nih.gov/books/NBK534829/',
    'Doxepin': 'https://www.ncbi.nlm.nih.gov/books/NBK542306/',
    'Duloxetine': 'https://www.ncbi.nlm.nih.gov/books/NBK549806/',
    'Escitalopram': 'https://www.ncbi.nlm.nih.gov/books/NBK557734/',
    'Fluoxetine': 'https://www.ncbi.nlm.nih.gov/books/NBK459223/',
    'Mirtazapine': 'https://www.ncbi.nlm.nih.gov/books/NBK519059/',
    'Nortriptyline': 'https://www.ncbi.nlm.nih.gov/books/NBK482214/',
    'Paroxetine': 'https://www.ncbi.nlm.nih.gov/books/NBK526022/',
    'Ropinirole': 'https://www.ncbi.nlm.nih.gov/books/NBK554532/',
    'Sertraline': 'https://www.ncbi.nlm.nih.gov/books/NBK547689/',
    'Trazodone': 'https://www.ncbi.nlm.nih.gov/books/NBK470560/',
    'Venlafaxine': 'https://www.ncbi.nlm.nih.gov/books/NBK548799/',
    'Less common antidepressants or combination of antidepressants': 'https://www.ncbi.nlm.nih.gov/books/NBK538182/'
}
# Adding a new column 'antidep_link' to the table based on 'antidep' column values
coefficients['antidep_link'] = coefficients['antidep'].map(nih_links)
```

```python
# Split the table by the 'antidep' column
groups = coefficients.groupby('antidep')

# Initialize variables
content = ""
coef_sum_dict = {group: 0 for group in groups.groups.keys()}

def calculate_coef_sum(strata, current_sum, group_df):
    # Get the coefficient for the given strata within the specific group
    coef = group_df.loc[group_df['strata'] == strata, 'coef']
    if not coef.empty:
        # Update the coefficient sum
        current_sum += coef.values[0]
    return current_sum

def process_content(content):
    # Split the content string into individual strata
    strata_list = content.split(" | ")
    # Dictionary to store coefficient sums for each group
    coef_sum_dict = {group: 0 for group in groups.groups.keys()}
    for strata in strata_list:
        for group, group_df in groups:
            coef_sum_dict[group] = calculate_coef_sum(strata, coef_sum_dict[group], group_df)
    # Calculate probabilities using the sigmoid function
    probability_dict = {group: 1 / (1 + np.exp(-coef_sum)) for group, coef_sum in coef_sum_dict.items()}
    # Return the highest probability
    return max(probability_dict.values()), probability_dict

while True:
    new_strata = input("Enter strata (or 'exit' to stop): ")
    if new_strata.lower() == 'exit':
        break
    if content:
        content += " | " + new_strata
    else:
        content = new_strata

    highest_probability, probability_dict = process_content(content)
    # Plot the pie chart
    response_prob = highest_probability * 100
    no_response_prob = 100 - response_prob
    
    # Bold text
    bold_start = "\033[1m"
    bold_end = "\033[0m"
    
    # Underline text
    underline_start = "\033[4m"
    underline_end = "\033[0m"

    # Print the report
    print(f"{bold_start}{underline_start}\nAntidepressant Recommendation Report\n{underline_end}{bold_end}")
    print(f"\n{bold_start}Relevant History\n{bold_end}")
    for item in content.split(" | "):
        print(f"- {item}")
    print("\nWe will use other patients' response experiences to recommend antidepressants to you based on your history.\n")
    
    print(f"\n{bold_start}Executive Summary\n{bold_end}")
    recommended_antidep = max(probability_dict, key=probability_dict.get)
    print(f"Among patients with a similar medical history as you, {recommended_antidep} gives you the highest chances for response. {response_prob:.2f}% of patients who took it experienced response.")

    # Data to plot
    labels = ['Response', 'No Response']
    sizes = [response_prob, no_response_prob]
    colors = ['seagreen', 'khaki']

    # Plot
    plt.figure(figsize=(3,3))
    plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 8}, radius=0.5)
    plt.axis('equal')
    plt.title(f"Figure 1: {recommended_antidep} is Best for You", fontsize=8)
    plt.show()
    
    print(f"\n{bold_start}Recommendation{bold_end}")
    print("\nResponse means that your symptoms will be reduced by 50%. We examined the experiences of more than 3 million patients and more than 10 million antidepressant treatments.")
    print(f"\nWe recommend you take {recommended_antidep}.\n")  
    
    # Create the bar chart
    sorted_probs = sorted(probability_dict.items(), key=lambda x: x[1], reverse=True)
    labels = [item[0] if item[0] != 'Less common antidepressants or combination of antidepressants' else 'Other' for item in sorted_probs]
    sizes = [item[1] * 100 for item in sorted_probs]
    
    plt.figure(figsize=(15, 8))
    bars = plt.barh(labels, sizes, color=colors)
    plt.xlabel('Response Rate (%)')
    plt.ylabel('Antidepressants')
    plt.title('Figure 2: Best Antidepressants for Patients with Similar Medical History')

    for bar, size in zip(bars, sizes):
        plt.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height()/2, f'{size:.1f}%', va='center')
    
    plt.gca().invert_yaxis()  # Invert y-axis to have the highest bar on top
    plt.show()
    
    # Additional information
    if 'antidep_link' in coefficients.columns:
        antidep_link = coefficients.loc[coefficients['antidep'] == recommended_antidep, 'antidep_link'].iloc[0]
        print(f"Learn more about Medications: {antidep_link}\n")
    else:
        print("Learn more about this medication: No link available\n")
    
    print("For more information on the effectiveness of common antidepressants, see: https://pubmed.ncbi.nlm.nih.gov/34877511/\n")
```
